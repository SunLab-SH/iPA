```python
file_dir = 'I:\\Bing\\fluorescence\\3D\\20230320MT-ISG\\04_rdf\\' 
file_dir2 = 'I:\\Bing\\fluorescence\\3D\\20230320MT-ISG\\00_index\\' 
output_dir = 'I:\\Bing\\fluorescence\\3D\\20230320MT-ISG\\06_stat\\'    # Modified
N_shell = 9

file_list=[]
with open(f'../file_list_MT.txt','r') as f:                           # Modified
	for line in f:
		file_list.append(line.strip('\n'))
                
            
isg_percent_total, mt_percent_total = [],  []
for file_name in file_list:
    print(file_name)

    with open(f'{file_dir}{file_name}_bin2_rdf.xvg') as fp:
        k1 = fp.readlines()

    N_isg = len(k1)/N_shell
    isg_total_vol = N_isg - k1[0 : int(N_isg*1)].count('\n')

    k1 = np.loadtxt(f'{file_dir2}{file_name}_bin2_ne_index.xvg')
    xs = [item[0] for item in k1]
    ys = [item[1] for item in k1]
    zs = [item[2] for item in k1]
    k2 = np.loadtxt(f'{file_dir2}{file_name}_bin2_pm_index.xvg')
    xs2 = [item[0] for item in k2]
    ys2 = [item[1] for item in k2]
    zs2 = [item[2] for item in k2]

    #print(xs)

    pm_edge = np.zeros((64, 800, 800))   # Modified
    ne_edge = np.zeros((64, 800, 800)) 

    for i in range(0, len(k2)):
        pm_edge[int(xs2[i]),int(ys2[i]),int(zs2[i])] = 1
    for i in range(0, len(k1)):
        ne_edge[int(xs[i]),int(ys[i]),int(zs[i])] = 1

    cell_total_vol = 0
    for z in range(64):
        # Create an empty mask with the same shape as the contour line array
        mask = np.zeros_like(pm_edge[z], dtype=np.uint8)
        # Find contours from the binary contour line array
        contours, _ = cv2.findContours(pm_edge[z].astype('uint8'), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        # Fill the interior of each contour with white color in the mask
        cv2.drawContours(mask, contours, -1, (255), cv2.FILLED, offset=(0, 0))
        mask_array = np.asarray(mask/255)

        mask_ne = np.zeros_like(ne_edge[z], dtype=np.uint8)
        contours_ne, _ = cv2.findContours(ne_edge[z].astype('uint8'), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        cv2.drawContours(mask_ne, contours_ne, -1, (255), cv2.FILLED, offset=(0, 0))
        mask_ne_array = np.asarray(mask_ne/255)

        mask_vol = np.sum(mask_array) - np.sum(mask_ne_array)
        cell_total_vol += mask_vol

    isg_percent = isg_total_vol/cell_total_vol

    img_mt = mrcfile.open(f'{output_dir}{file_name}_volumn_bin2_MT.mrc', mode='r').data
    
    MT_total_vol = 0
    for z in range(img_mt.shape[0]):
        thresh = filters.threshold_otsu(img_mt[z]) 
        MT_voxel = morphology.closing(img_mt[z] > thresh, morphology.square(3))
        MT_voxels = np.sum(MT_voxel)
        MT_total_vol += MT_voxels

    mt_percent = MT_total_vol/cell_total_vol

    isg_percent_total.append(isg_percent)
    mt_percent_total.append(mt_percent)

    print(isg_percent, mt_percent)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
# Replace these with your actual data
group_names = ['Group 1', 'Group 2', 'Group 3']
data_without_drug = np.random.randn(3)
data_with_drug = np.random.randn(3)

# Calculate mean and standard error of the mean (SEM)
mean_without_drug = np.mean(data_without_drug)
sem_without_drug = np.std(data_without_drug) / np.sqrt(len(data_without_drug))
mean_with_drug = np.mean(data_with_drug)
sem_with_drug = np.std(data_with_drug) / np.sqrt(len(data_with_drug))

# Plot the data points
plt.scatter(np.arange(len(data_without_drug)), data_without_drug, label='Without Drug')
plt.scatter(np.arange(len(data_with_drug)), data_with_drug, label='With Drug')

# Plot the error bars
plt.errorbar(np.arange(len(data_without_drug)), data_without_drug, yerr=sem_without_drug, linestyle='None', color='black', capsize=5)
plt.errorbar(np.arange(len(data_with_drug)), data_with_drug, yerr=sem_with_drug, linestyle='None', color='black', capsize=5)

# Plot the mean values as a line on top of the bar
plt.plot([-0.2, 0.2], [mean_without_drug, mean_without_drug], color='red', linewidth=2)
plt.plot([0.8, 1.2], [mean_with_drug, mean_with_drug], color='red', linewidth=2)

# Customize the plot
plt.xticks(range(len(group_names)), group_names)
plt.xlabel('Groups')
plt.ylabel('Data')
plt.title('Data Comparison: With vs Without Drug')
plt.legend()

# Display the plot
plt.show()
```
